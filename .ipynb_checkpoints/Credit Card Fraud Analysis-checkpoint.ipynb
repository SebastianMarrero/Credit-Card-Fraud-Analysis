# Credit Card Fraud Analysis

##### The following data analysis will examine and model synthetic credit card consumer data produced for the 2022 Conference on Neural Information Processing Systems. the dataset contains fraudulent transactions which our analysis will attempt to model using classification modeling, additionally incorporating visualizations to clarify understanding of the data. Built with Python, Scikit-learn, Matplotlib, and Seaborn, the project includes data visualizations and performance metrics (e.g., AUC-ROC) to enhance interpretability. The objective is to simulate real-world fraud detection scenarios for financial risk mitigation.

Each instance in the dataset represents a feature-engineered, synthetic bank account application generated by a deep learning model titled Conditional Tabular Generative Adversarial Network (CTGAN) trained on real-world anonymized bank application data. Dataset derived from Kaggle Bank Account Fraud Dataset Suite (NeurIPS 2022) (https://www.kaggle.com/datasets/sgpjesus/bank-account-fraud-dataset-neurips-2022?select=Base.csv).
- Dataset suite Github homepage: https://github.com/feedzai/bank-account-fraud/tree/main

### Scaling and Modeling 
- Our data is highly imbalanced, with only 1% of the data containing the fraud class.
- Our modeling analysis will consist of a comparison between XGBoost / LightGBM and Logistic Regression, additionally incorporating Oversampling techniques and meta-modeling techniques such as SMOTE/ADASYN and Ensembles & Stacking to enrich the comparitive analysis and the ensuing predictive outcomes

- The below code will split our data into training and test sets and impute Nan Values in our columns with median values to preserve consistency and integrity of data and to comply with copying our data for our oversampling technique.

```python
from sklearn.model_selection import train_test_split
CreditCardDfM=CreditCardDf.drop(columns='fraud_bool')

## Splitting the Data
x_train, x_test, y_train, y_test = train_test_split(CreditCardDfM,
                                                    CreditCardDf['fraud_bool'],test_size=0.2,
                                                    stratify=CreditCardDf['fraud_bool'], random_state=42)

## Must split data into numerical and categorical features to impute values correctly
CreditCardDfMNum=CreditCardDfM.select_dtypes(include='number').columns
CreditCardDfMCate=CreditCardDfM.select_dtypes(include='object').columns

from sklearn.impute import SimpleImputer
imputer_num=SimpleImputer(strategy='median')
x_train[CreditCardDfMNum] = imputer_num.fit_transform(x_train[CreditCardDfMNum])
x_test[CreditCardDfMNum] = imputer_num.fit_transform(x_test[CreditCardDfMNum])

imputer_cate=SimpleImputer(strategy='most_frequent')
x_train[CreditCardDfMCate] = imputer_cate.fit_transform(x_train[CreditCardDfMCate])
x_test[CreditCardDfMCate] = imputer_cate.fit_transform(x_test[CreditCardDfMCate])
```

```python
## Standard Scaling 
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
x_train[CreditCardDfMNum] = scaler.fit_transform(x_train[CreditCardDfMNum])
x_test[CreditCardDfMNum] = scaler.fit_transform(x_test[CreditCardDfMNum])
```

```python
## Label Encoding
from sklearn.preprocessing import OneHotEncoder
encoder = OneHotEncoder(handle_unknown='ignore', sparse_output=False)

x_train_cat = encoder.fit_transform(x_train[CreditCardDfMCate])
x_test_cat = encoder.transform(x_test[CreditCardDfMCate])

x_train_num = x_train.drop(columns=CreditCardDfMCate).values
x_test_num = x_test.drop(columns=CreditCardDfMCate).values

x_train = np.hstack([x_train_num, x_train_cat])
x_test = np.hstack([x_test_num, x_test_cat])
```

---
- Creating a second set of training data to apply SMOTE oversampling technique

```python
x_train_og=x_train.copy()
y_train_og=y_train.copy()

from imblearn.over_sampling import SMOTE
smote = SMOTE(random_state=42)
x_train_smote, y_train_smote = smote.fit_resample(x_train, y_train)
```

----
## Logistic Regression
##### With and Without SMOTE-altered data

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import classification_report, confusion_matrix, precision_recall_curve, auc

## Training both models
log_reg_orig = LogisticRegression(max_iter=1000, random_state=42)
log_reg_orig.fit(x_train_og,y_train_og)

log_reg_smote = LogisticRegression(max_iter=1000, random_state=42)
log_reg_smote.fit(x_train_smote,y_train_smote)
```

```python
## Predictions
y_pred_orig = log_reg_orig.predict(x_test)
y_pred_smote = log_reg_smote.predict(x_test)
```

```python
## Evaluating and comparing the models 
print("Logistic Regression with Original Training Data:")
print(confusion_matrix(y_test, y_pred_orig))
print(classification_report(y_test, y_pred_orig))

print("Logistic Regression with SMOTE-altered Training Data:")
print(confusion_matrix(y_test, y_pred_smote))
print(classification_report(y_test, y_pred_smote))
```

#### Let's discuss the results here:
---
Our Logistic Regression with Original Training data:
- Correctly identified almost all non-fraud applications
- Only misclassified 17 applications as frauds
- mistook 2176 frauds for regular applications
- Only predicted 30 fraudulent applications

Out of all fraud applications, 64% of the predictions were correct (recall, only 30 applications). Recall is at an abysmal 0.01, meaning we caught only 1% of fraudulent applications. F-1 score indicates a terrible balance of precision and recall ~ 0.03. Overall accuracy, at 0.99, is misleading since our data is inherently significantly skewed.

--- 
Our Logistic Regression with SMOTE-altered Training data:
- Correctly identified less non-fraudulent applications than our LR with original, non-SMOTE data
- Missed 514 fraudulent applications
- Predicted over 1692 fraudulent applications, a huge step-up from our original training data
- Predicted fraud for 35219 non-fraudulent applications

Out of all fraud applications, 5% of the predictions were correct (recall, 1692 applications). Recall is at a much improved 0.77, meaning we caught up to 77% of fraudulent applications. F-1 score indicates a subpar balance of precision and recall ~ 0.09. Overall accuracy, at 0.82, is underwhelming and worse-off than our logistic regression completed with non-SMOTE training data. Our Smote data allowed our model to catch much more fraudulent transactions, however its casting a much, much wider net, resulting in false positives.

----


### Precision-Recall curves for Logistic Regression Visualization

```python
import matplotlib.pyplot as plt

def precision_plot(model,x_test,y_test,label):
    y_scores = model.predict_proba(x_test)[:, 1]
    precision, recall, _ = precision_recall_curve(y_test, y_scores)
    pr_auc = auc(recall, precision)
    plt.plot(recall, precision, label=f'{label} (PR AUC = {pr_auc:.3f})')

plt.figure(figsize=(8, 6))
precision_plot(log_reg_orig, x_test, y_test, "LogReg Original")
precision_plot(log_reg_smote, x_test, y_test, "LogReg SMOTE")
plt.xlabel("Recall")
plt.ylabel("Precision")
plt.title("Precision-Recall Curve - Logistic Regression")
plt.legend()
plt.grid()
plt.tight_layout()
plt.savefig('assets/images/Precision-Recall_Curve_LR.png')
plt.show()
```

What do our findings show? 

- The LogReg Original model (blue curve) shows a slightly higher PR AUC (0.140), confirming its strong precision from the confusion matrix, where only 17 false positives were made (35219 FP were made in the SMOTE regression).

- The LogReg SMOTE model (orange curve) reaches much further along the recall axis, which aligns with it catching 1,692 fraud cases compared to just 30 in the original model.

- The sharp decline in the SMOTE modelâ€™s precision, indicated by the lower curve, is consistent with its high number of false positives and low precision score of 0.05.

- Both curves drop quickly as recall increases, showing that the models struggle to maintain precision when trying to identify more fraud cases.

The overall shape of the curves validates the classification reports: the original model is conservative and precise, while the SMOTE model is more aggressive in catching fraud but generates many false alarms In general, SMOTE improves fraud detection recall, but significantly worsens precision â€” meaning your model is better at catching frauds, but triggers a lot of false alarms in the process.

## Extreme Gradient Boosting (XGBoost) analysis
##### XGBoost and LightGBM are two of the most powerful models for tabular classification for data like this one

```python
## Below code will label encode our categorical values, replacing those strings with integer values 
## Tree-based models such as XGBoost and LightGBM care about splits in data, not numeric distances (as was the case with One Hot encoding)

# We'll need to recover and re-run our pre-OneHotEncoding x_train arrays to bypass errors associated with running them through two encoding techniques
from sklearn.preprocessing import LabelEncoder

x_train, x_test, y_train, y_test = train_test_split(CreditCardDfM,
                                                    CreditCardDf['fraud_bool'],test_size=0.2,
                                                    stratify=CreditCardDf['fraud_bool'], random_state=42)

CreditCardDfMNum=CreditCardDfM.select_dtypes(include='number').columns
CreditCardDfMCate=CreditCardDfM.select_dtypes(include='object').columns

imputer_num=SimpleImputer(strategy='median')
x_train[CreditCardDfMNum] = imputer_num.fit_transform(x_train[CreditCardDfMNum])
x_test[CreditCardDfMNum] = imputer_num.transform(x_test[CreditCardDfMNum])

imputer_cate=SimpleImputer(strategy='most_frequent')
x_train[CreditCardDfMCate] = imputer_cate.fit_transform(x_train[CreditCardDfMCate])
x_test[CreditCardDfMCate] = imputer_cate.fit_transform(x_test[CreditCardDfMCate])

scaler = StandardScaler()
x_train[CreditCardDfMNum] = scaler.fit_transform(x_train[CreditCardDfMNum])
x_test[CreditCardDfMNum] = scaler.fit_transform(x_test[CreditCardDfMNum])

for col in CreditCardDfMCate:
    le = LabelEncoder()
    x_train[col] = le.fit_transform(x_train[col])
    x_test[col] = le.transform(x_test[col])


x_train_og=x_train.copy()
y_train_og=y_train.copy()

from imblearn.over_sampling import SMOTE
smote = SMOTE(random_state=42)
x_train_smote, y_train_smote = smote.fit_resample(x_train, y_train)
```

```python
# The following code will train xgboost on our non-smote, imbalanced data
import xgboost as xgb
from sklearn.metrics import classification_report, confusion_matrix, precision_recall_curve, roc_auc_score, auc, roc_curve

xgb_model_orig=xgb.XGBClassifier(Use_label_encoder=False, eval_metric='logloss',random_state=42)
xgb_model_orig.fit(x_train_og,y_train_og)

# The following code will train xgboost on our smote, balanced data
xgb_model_smote=xgb.XGBClassifier(Use_label_encoder=False, eval_metric='logloss',random_state=42)
xgb_model_smote.fit(x_train_smote,y_train_smote)
```

```python
y_pred_xgb_orig = xgb_model_orig.predict(x_test)
y_score_xgb_orig = xgb_model_orig.predict_proba(x_test)[:,1]

y_pred_xgb_smote = xgb_model_smote.predict(x_test)
y_score_xgb_smote = xgb_model_smote.predict_proba(x_test)[:,1]

print("XGBoost with original, imbalanced data:")
print(confusion_matrix(y_test, y_pred_xgb_orig))
print(classification_report(y_test, y_pred_xgb_orig))

print("XGBoost with Smote-applied, balanced data:")
print(confusion_matrix(y_test, y_pred_xgb_smote))
print(classification_report(y_test, y_pred_xgb_smote))
```

#### Let's discuss the results here:
---
XGBoost with Original, Imbalanced Training Data:
- Correctly identified 197,687 out of 197,794 non-fraud cases (very high precision on the majority class).

- Incorrectly flagged 107 non-fraud cases as fraud (slightly more false positives than Logistic Regression).

- Missed 2,133 out of 2,206 fraud cases â€” meaning only 73 frauds were caught.

- Precision for fraud predictions was 41%, but recall was just 3%.

Although precision for the fraud class looks better than in Logistic Regression, recall remains extremely low â€” the model still missed 97% of fraud. The f1-score is low (0.06), indicating poor fraud detection. Accuracy remains high at 0.99, but again, this is misleading due to the dataset imbalance.

--- 

XGBoost with SMOTE-balanced Training Data:
- Correctly flagged 1,996 out of 2,206 fraud cases â€” a huge recall boost (90%).

- Mistakenly labeled 130,066 non-fraud cases as fraud, a dramatic increase in false positives.

- Precision on fraud predictions dropped to just 2%.

- Overall accuracy plummeted to 35% due to overwhelming misclassification of legitimate transactions.

This version of XGBoost is extremely sensitive to fraud â€” catching almost all of it â€” but at the cost of flagging the majority of normal transactions incorrectly. The model has high recall but extremely poor precision and overall usability, reflecting an overly aggressive fraud detection behavior after SMOTE oversampling.

```python
def XGBoost_PR_curve(y_true, y_scores, label):
    precision, recall, _ = precision_recall_curve(y_true, y_scores)
    pr_auc = auc(recall, precision)
    plt.plot(recall, precision, label=f"{label} (AUC = {pr_auc:.3f})")

plt.figure(figsize=(8, 6))

# Curve for XGBoost - original training data
XGBoost_PR_curve(y_test, y_score_xgb_orig, "XGBoost (Original)")

# Curve for XGBoost - SMOTE training data
XGBoost_PR_curve(y_test, y_score_xgb_smote, "XGBoost (SMOTE)")

# Plot settings
plt.xlabel("Recall")
plt.ylabel("Precision")
plt.title("Precision-Recall Curve - XGBoost Comparison")
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.savefig("assets/images/Precision-Recall Curve_XGBoost")
plt.show()
```

What do our findings show? 
- XGBoost (Original) has a PR AUC of 0.118, significantly outperforming the SMOTE model (PR AUC of 0.031).
- The original model maintains higher precision across almost all levels of recall, indicating that it makes fewer false-positive fraud predictions.
- XGBoost (SMOTE) achieves higher recall briefly at the very start, but its precision drops rapidly and stays near zero, meaning it's flagging a massive number of non-fraud cases as fraud.
- The flat shape of the SMOTE curve confirms its poor precision and overall unreliability, despite high sensitivity to fraud.

- XGBoost trained on the original data performs significantly better than XGBoost trained on the balanced data â€” it maintains a better balance between catching fraud and not overwhelming the system with false positives.

- SMOTE hurts XGBoost performance here, likely due to overfitting on synthetic samples or misrepresenting class boundaries.

The original model is more conservative, yet far more trustworthy and usable for fraud detection in real-world applications.

#### Why did SMOTE hurt our XGBoost model so much?
- Synthetic points, as in the case of our oversampling technique in SMOTE, adds a lot of "noise" in the form of newly generated fraud samples, which can mislead the tree growth process.
- Our XGBoost model is incredibly vulnerable to synthetic patterns generated by SMOTE due to it being a high-capacity learner - these patterns don't generalize uniformly to the rest of the data, leading to a precipitous drop in precision score.

### Importantly: 
This catastrophic drop in precision is a textbook case of tree models overfitting to synthetic patterns introduced by SMOTE. Because LightGBM and XGBoost build decision splits on exact feature values, the interpolated samples from SMOTE inaccurately reflect the true structure of the data. Instead, they introduce unnatural boundaries the model vulnerably focuses on â€” allowing it to memorize synthetic "fraud zones" that donâ€™t generalize, leading to massive false positive rates.

```python
def LightGBM_PR_Curve(y_true, y_scores, label):
    precision, recall, _ = precision_recall_curve(y_true, y_scores)
    pr_auc = auc(recall, precision)
    plt.plot(recall, precision, label=f"{label} (AUC = {pr_auc:.3f})")

plt.figure(figsize=(8, 6))

LightGBM_PR_Curve(y_test, y_score_lgb_orig, "LightGBM (Original)")

LightGBM_PR_Curve(y_test, y_score_lgb_smote, "LightGBM (SMOTE)")

plt.xlabel("Recall")
plt.ylabel("Precision")
plt.title("Precision-Recall Curve - LightGBM Comparison")
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.savefig("assets/images/LightGBM_PR_Curve.png")
plt.show()
```

What do our findings show?

- Despite low recall, our model trained on imbalanced data retains predictive value â€” it makes some smart decisions in identifying true fraud.
- Our SMOTE Model's precision hovers near zero across all recall levels, meaning that almost all fraud predictions are false alarms.
- The curve confirms what we saw in the classification report: SMOTE completely destroyed LightGBMâ€™s ability to distinguish legitimate from fraudulent behavior.

## Adding scale_pos_weight for LightGBM and XGBoost for proper fraud class weighting
##### Following Our failure to produce dependable models with SMOTE, scale_pos_weight will ensure our fraud data is properly weighted 


```python
## the following code will estimate the relative weight of fraud cases compared to non-fraud casesin our data set
neg = sum(y_train == 0)
pos = sum(y_train == 1)
scale_pos_weight = neg / pos
```

```python
## The following code will re-run our XGBoost and LightGBM models, however we will disregard the SMOTE data and implement the scale_pos_weight parameter
xgb_weighted=xgb.XGBClassifier(scale_pos_weight=neg/pos, Use_label_encoder=False, eval_metric='logloss',random_state=42)
xgb_weighted.fit(x_train_og,y_train_og)

y_pred_xgb_weighted = xgb_weighted.predict(x_test)
y_score_xgb_weighted = xgb_weighted.predict_proba(x_test)[:,1]

print("XGBoost with scale_pos_weight")
print(confusion_matrix(y_test, y_pred_xgb_weighted))
print(classification_report(y_test, y_pred_xgb_weighted))

lgb_weighted = lgb.LGBMClassifier(scale_pos_weight=neg/pos, random_state=42)
lgb_weighted.fit(x_train_og, y_train_og) 

y_pred_lgb_weighted = lgb_weighted.predict(x_test)
y_score_lgb_weighted = lgb_weighted.predict_proba(x_test)[:,1]

print("LightGBM -  scale_pos_weight")
print(confusion_matrix(y_test, y_pred_lgb_weighted))
print(classification_report(y_test, y_pred_lgb_weighted))
```

```python
def Weighted_PR_Curve(y_true, y_scores, label):
    precision, recall, _ = precision_recall_curve(y_true, y_scores)
    pr_auc = auc(recall, precision)
    plt.plot(recall, precision, label=f"{label} (AUC = {pr_auc:.3f})")

plt.figure(figsize=(8, 6))

Weighted_PR_Curve(y_test, y_score_xgb_weighted, "XGBoost - Weighted")
Weighted_PR_Curve(y_test, y_score_lgb_weighted, "LightGBM - Weighted")

plt.xlabel("Recall")
plt.ylabel("Precision")
plt.title("Precision-Recall Curve - Weighted LightGBM/XGBoost Comparison")
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.savefig("assets/images/Weighted_PR_Curve.png")
plt.show()
```
